```python
import pandas as pd
```

```python
cs_publications = pd.read_csv('CS_publications.csv')
qtm_publications = pd.read_csv('QTM_publications.csv')

cs_authors = pd.read_csv('CS_authors.csv')
qtm_authors = pd.read_csv('QTM_authors.csv')
```

```python
publications = pd.concat([cs_publications, qtm_publications], ignore_index=True)
authors = pd.concat([cs_authors, qtm_authors], ignore_index=True)

print(publications.shape)
```

```python
merged = pd.merge(publications, authors, left_on='Author', right_on='Name', how='left')

merged = merged.reset_index()
display(merged)
```

```python
merged = merged.drop(columns=['Name'])
merged.to_csv('merged_publications.csv', index=False)
```
